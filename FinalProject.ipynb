# Black or Colored (HAH)

```python
def equal_black(a):
    if (a.red == 0) and (a.green == 0) and (a.blue == 0):             
        return True
    else:
        return False
    #Checks if cell is empty or not

def equal_color(a,b):
    if (a.red,a.green,a.blue)==(b.red,b.green,b.blue):       
        return True 
    else:
        return False
    #Compares the colors the two blocks
```

# Satisfaction

```python
def satisfaction(grid,i,j):
    if grid[i,j]!=(0,0,0):      
        GridSize = grid.width
        sat = 0                  
        
##############################################################################################################################       

        #Edges    
        if i == 0 or j == 0 or i == GridSize-1 or j == GridSize-1:   
            MaxNeighbors = 5
            
            #Top Leff #YOU CHANGED THE J+1 TO J-1 
            if i == 0 and j == 0:    
                MaxNeighbors = 3     

                #One Right
                if equal_color(grid[i+1,j],grid[i,j]):  
                    sat = sat + 1                       
                elif equal_black(grid[i+1,j]):           
                    MaxNeighbors -= 1                   
               
                #One Right, One Below
                if equal_color(grid[i+1,j+1],grid[i,j]): 
                    sat = sat + 1 
                elif equal_black(grid[i+1,j+1]):         
                    MaxNeighbors -= 1
                
                #One Below
                if equal_color(grid[i,j+1],grid[i,j]):   
                    sat = sat + 1
                elif equal_black(grid[i,j+1]):
                    MaxNeighbors -= 1
                
                #Is it Satisfied?
                if MaxNeighbors == 0:  
                    satisfaction = 0    
                    #otherwise we divide by zero and we don't want that
                    return satisfaction
                else:
                    satisfaction = float(sat) / float(MaxNeighbors) 
                    return satisfaction    

###################################################################################################                

        #Bottom Leff  #YOU CHANGED J+1 TO I+1 AND J-1 TO J+1  
            elif i == 0 and j == GridSize-1:       
                MaxNeighbors = 3

                #One Above
                if equal_color(grid[i,j-1],grid[i,j]):  
                    sat = sat + 1
                elif equal_black(grid[i,j-1]):
                    MaxNeighbors -= 1

                #One Right, One Above
                if equal_color(grid[i+1,j-1],grid[i,j]): 
                    sat = sat + 1
                elif equal_black(grid[i+1,j-1]):
                    MaxNeighbors -= 1

                #One Right
                if equal_color(grid[i+1,j],grid[i,j]):
                    sat = sat + 1
                elif equal_black(grid[i+1,j]):
                    MaxNeighbors -= 1

                #Is it Satisfied?    
                if MaxNeighbors == 0:
                    satisfaction = 0
                    return satisfaction
                else:
                    satisfaction = float(sat) / float(MaxNeighbors)
                    return satisfaction

###################################################################################################

        #Top Right #YOU CHANGED J+1 TO J-1   
            elif (i==GridSize-1 and j ==0):     
                MaxNeighbors = 3

                #One Left
                if equal_color(grid[i-1,j],grid[i,j]):
                    sat = sat + 1
                elif equal_black(grid[i-1,j]):
                    MaxNeighbors -= 1

                #One Left, One Below
                if equal_color(grid[i-1,j+1],grid[i,j]):
                    sat = sat + 1
                elif equal_black(grid[i-1,j+1]):
                    MaxNeighbors -= 1

                #One Below  
                if equal_color(grid[i,j+1],grid[i,j]):
                    sat = sat + 1
                elif equal_black(grid[i,j+1]):
                    MaxNeighbors -= 1

                #Is it Satisied?
                if MaxNeighbors == 0:
                    satisfaction = 0
                    return satisfaction
                else:
                    satisfaction = float(sat) / float(MaxNeighbors)
                    return satisfaction
            
######################################################################################################
            
            #Bottom Right
            elif (i == GridSize-1  and j == GridSize-1):   
                MaxNeighbors = 3
                
                #One Left
                if equal_color(grid[i-1,j],grid[i,j]):
                    sat = sat + 1
                elif equal_black(grid[i-1,j]):
                    MaxNeighbors -= 1
                
                #One Left, One Up    
                if equal_color(grid[i-1,j-1],grid[i,j]):
                    sat = sat + 1
                elif equal_black(grid[i-1,j-1]):
                    MaxNeighbors -= 1
                
                #One Up
                if equal_color(grid[i,j-1],grid[i,j]):
                    sat = sat + 1
                elif equal_black(grid[i,j-1]):
                    MaxNeighbors -= 1

                #Is it Satisfied?
                if MaxNeighbors == 0:
                    satisfaction = 0
                    return satisfaction
                else:
                    satisfaction = float(sat) / float(MaxNeighbors)
                    return satisfaction
                
########################################################################################################
########################################################################################################
  
    #EDGES!!! Max Neighbors is now 5
            
        #Leff Edge
        elif i == 0 and j != 0 and j!=GridSize-1 :  

            #One Right
            if equal_color(grid[i+1,j],grid[i,j]):
                sat = sat + 1
            elif equal_black(grid[i+1,j]):
                MaxNeighbors -= 1

            #One Right, One Up
            if equal_color(grid[i+1,j-1],grid[i,j]):
                sat = sat + 1
            elif equal_black(grid[i+1,j-1]):
                MaxNeighbors -= 1

            #One Right, One Down
            if equal_color(grid[i+1,j+1],grid[i,j]): 
                sat = sat + 1
            elif equal_black(grid[i+1,j+1]):
                MaxNeighbors -= 1

            #One Down
            if equal_color(grid[i,j+1],grid[i,j]):
                sat = sat + 1
            elif equal_black(grid[i,j+1]):
                MaxNeighbors -= 1

            #One Up
            if equal_color(grid[i,j-1],grid[i,j]):
                sat = sat + 1
            elif equal_black(grid[i,j-1]):
                MaxNeighbors -= 1

            #Is it Satisfied
            if MaxNeighbors == 0:
                satisfaction = 0
                return satisfaction
            else:
                satisfaction = float(sat) / float(MaxNeighbors)
                return satisfaction
            
###############################################################################################################
        
        #Top Edge
        elif j == 0 and i!=0 and i!=GridSize-1:  
            
            #One Right
            if equal_color(grid[i+1,j],grid[i,j]):
                sat = sat + 1
            elif equal_black(grid[i+1,j]):
                MaxNeighbors -= 1
            
            #One Right, One Down
            if equal_color(grid[i+1,j+1],grid[i,j]):
                sat = sat + 1
            elif equal_black(grid[i+1,j+1]):
                MaxNeighbors -= 1
            
            #One Down
            if equal_color(grid[i,j+1],grid[i,j]):
                sat = sat + 1
            elif equal_black(grid[i,j+1]):
                MaxNeighbors -= 1
            
            #One Leff, One Down
            if equal_color(grid[i-1,j+1],grid[i,j]):
                sat = sat + 1
            elif equal_black(grid[i-1,j+1]):
                MaxNeighbors -= 1
            
            #One Leff
            if equal_color(grid[i-1,j],grid[i,j]):
                sat = sat + 1
            elif equal_black(grid[i-1,j]):
                MaxNeighbors -= 1

            #Is it Satisfied?
            if MaxNeighbors == 0:
                satisfaction = 0
                return satisfaction
            else:
                satisfaction = float(sat) / float(MaxNeighbors)
                return satisfaction
            
#################################################################################################################

        #Right Edge
        elif i == GridSize-1 and j !=0 and j!=GridSize-1:   
            
            #One Up
            if equal_color(grid[i,j-1],grid[i,j]):
                sat = sat + 1
            elif equal_black(grid[i,j-1]):
                MaxNeighbors -= 1
            
            #One Left, One Down
            if equal_color(grid[i-1, j+1],grid[i,j]):
                sat = sat + 1
            elif equal_black(grid[i-1, j+1]):
                MaxNeighbors -= 1
                
            #One Left
            if equal_color(grid[i-1,j],grid[i,j]):
                sat = sat + 1
            elif equal_black(grid[i-1,j]):
                MaxNeighbors -= 1
                
            #One Down
            if equal_color(grid[i,j+1],grid[i,j]):
                sat = sat + 1
            elif equal_black(grid[i,j+1]):
                MaxNeighbors -= 1
                
            #One Left, One Up
            if equal_color(grid[i-1,j-1],grid[i,j]):
                sat = sat + 1
            elif equal_black(grid[i-1,j-1]):
                MaxNeighbors -= 1

            #Is it Satisfied?
            if MaxNeighbors == 0:
                satisfaction = 0
                return satisfaction
            else:
                satisfaction = float(sat) / float(MaxNeighbors)
                return satisfaction
            
#################################################################################################################

        #Bottom Edge
        elif j == GridSize-1 and i!=0 and i != GridSize-1:  
            
            #One Right
            if equal_color(grid[i+1,j],grid[i,j]):
                sat = sat + 1
            elif equal_black(grid[i+1,j]):
                MaxNeighbors -= 1
                
            #One Right, One Up
            if equal_color(grid[i+1, j-1],grid[i,j]):
                sat = sat + 1
            elif equal_black(grid[i+1, j-1]):
                MaxNeighbors -= 1
                
            #One Leff
            if equal_color(grid[i-1,j],grid[i,j]):
                sat = sat + 1
            elif equal_black(grid[i-1,j]):
                MaxNeighbors -= 1
                
            #One Up
            if equal_color(grid[i,j-1],grid[i,j]):
                sat = sat + 1
            elif equal_black(grid[i,j-1]):
                MaxNeighbors -= 1
            
            #One Leff, One Up
            if equal_color(grid[i-1,j-1],grid[i,j]): 
                sat = sat + 1
            elif equal_black(grid[i-1,j-1]):
                MaxNeighbors -= 1

            #Is it Satisfied?
            if MaxNeighbors == 0:
                satisfaction = 0
                return satisfaction
            else:
                satisfaction = float(sat) / float(MaxNeighbors)
                return satisfaction

#########################################################################################################
#########################################################################################################

        #Inside!!! Max Neighbors is 8
        else:
            MaxNeighbors = 8
            
            #One Right
            if equal_color(grid[i+1,j],grid[i,j]):
                sat = sat + 1
            elif equal_black(grid[i+1,j],):
                MaxNeighbors -= 1
            
            #One Right, One Up
            if equal_color(grid[i+1,j-1],grid[i,j]):
                sat = sat + 1
            elif equal_black(grid[i+1,j-1]):
                MaxNeighbors -= 1
            
            #One Up
            if equal_color(grid[i,j-1],grid[i,j]):
                sat = sat + 1
            elif equal_black(grid[i,j-1]):
                MaxNeighbors -= 1
                
            #One Leff, One Down
            if equal_color(grid[i-1, j+1],grid[i,j]):
                sat = sat + 1
            elif equal_black(grid[i-1, j+1]):
                MaxNeighbors -= 1
                
            #One Leff
            if equal_color(grid[i-1,j],grid[i,j]):
                sat = sat + 1
            elif equal_black(grid[i-1,j]):
                MaxNeighbors -= 1
                
            #One Down
            if equal_color(grid[i,j+1],grid[i,j]):
                sat = sat + 1
            elif equal_black(grid[i,j+1]):
                MaxNeighbors -= 1
                
            #One Leff, One Up
            if equal_color(grid[i-1,j-1],grid[i,j]):
                sat = sat + 1
            elif equal_black(grid[i-1,j-1]):
                MaxNeighbors -= 1
                
            #One Right, One Down
            if equal_color(grid[i+1,j+1],grid[i,j]): 
                sat = sat + 1
            elif equal_black(grid[i+1,j+1]):
                MaxNeighbors -= 1
            
            #Is is Satisfied?
            if MaxNeighbors == 0:
                satisfaction = 0
                return satisfaction
            else:
                satisfaction = float(sat) / float(MaxNeighbors)
                return satisfaction

    else:            
        return None
```

# Finding and moving movers 

```python
def find_movers(grid,threshold):                              
    movers = []
    emptys = []
    for i in range(grid.width):             
        for j in range(grid.height):
            
            #Establishing black cells as empty
            if equal_black(grid[i,j]):      
                emptys.append((i,j))         
            
            # If cells are not satisfied, they become a mover
            else:                           
                if satisfaction(grid,i,j) < threshold:  
                    movers.append((i,j))                
    
    #Move mover cells into a random emmpty cell, changes color of empty cell
    for i in range(len(movers)-1):            
        if len(emptys) != 0:                    
            r = random.randint(0,len(emptys)-1) 
            grid[emptys[r]].rgb = grid[movers[i]].rgb 
            grid[movers[i]].rgb = (0,0,0)   
            emptys.remove(emptys[r])          
       
    clear_output()            
    grid.show()
```

```python
def Schelling(GridSize=5,Threshold=0.5,RedPop=60,BluePop=30,Iterations=1):
    grid = BlockGrid(GridSize,GridSize)           

    if (RedPop + BluePop)> 100:                    
        print "Choose values of (red plus blue) that don't exceed 100 %"
    
    else:
        for i in range(grid.height):                  
            for j in range(grid.width):              
                rand_int1 = random.randint(1,100)    
                
                if rand_int1 <= BluePop:             
                    grid[i,j] = (0, 0, 255)          
                
                elif rand_int1 <= (RedPop + BluePop):  
                    grid[i,j] = (255, 0, 0)          
                
                elif rand_int1 > (RedPop + BluePop): 
                    grid[i,j] = (0, 0, 0)            
        grid.show()                                    
                                                        
    while Iterations > 0:            
        find_movers(grid,Threshold)  
        Iterations = Iterations - 1  
v = interact(Schelling,GridSize=(5,100,5), Threshold=(0,1,.05), RedPop = (0,100,5), BluePop = (0,100,5) , Iterations = (0,400,5))#CODE FOR SLIDERS
```
